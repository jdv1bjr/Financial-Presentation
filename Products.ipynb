```python
types_of_products = merged_df['Category_Name'].value_counts()
types_of_products.head(10)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.barplot(x=types_of_products.index, y=types_of_products.values, color=sns.color_palette()[5])
plt.title('Types of Products')
plt.xlabel('Category')
plt.ylabel('Count')
plt.xticks(rotation=90)
plt.show()
```

```python
most_popular_products = merged_df['Product_Name'].value_counts()
most_popular_products.head(5).plot(kind='pie', subplots=True, figsize=(10, 10), autopct='%1.1f%%')
plt.title('Most Popular Products')
plt.xlabel('Product')
plt.ylabel('Count')
plt.xticks(rotation=90)
plt.show()
```
